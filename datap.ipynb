```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('static/download/data.csv')

per_tahun = df.groupby('tahun')['jumlah_sampah'].sum().reset_index()

per_tahun

plt.figure(figsize=(10,5))
plt.plot(per_tahun['tahun'], per_tahun['jumlah_sampah'], marker='o')  


plt.title("Jumlah Sampah per Tahun di Jawa Barat")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Sampah (Ton/Hari)")


plt.grid(alpha=0.3)

plt.show()
```
