```python
import pandas as pd

data = {'Nama': ['John', 'Jane', 'Bob', 'Alice'],
        'Usia': [25, 35, 30, 28],
        'Gaji': [50000, 60000, 70000, 55000]}

df = pd.DataFrame(data)
df
```

Gunakan loop for dan fungsi lambda untuk menghitung gaji setiap karyawan setelah diberikan peningkatan sebesar 5% dari gaji saat ini.

```python
hasil_tamp=[]
for i in data['Gaji']:
    naik_gaji = lambda i: i + (0.05) * i
    i=naik_gaji(i)
    hasil_tamp.append(i)

data_jumlah= {'Nama': ['John', 'Jane', 'Bob', 'Alice'],
            'Gajih_baru' :hasil_tamp} 
df_baru= pd.DataFrame(data_jumlah)
jumlah= df_baru['Gajih_baru'].sum()

data_jumlah= {'Usia': [25, 35, 30, 28],
            'Nama': ['John', 'Jane', 'Bob', 'Alice'],
            'Gajih_baru' :hasil_tamp,
            'Total_Gaji': [jumlah,jumlah,jumlah,jumlah]}
data_jumlah =pd.DataFrame(data_jumlah)
data_jumlah
```

Gunakan loop for lagi untuk mengevaluasi karyawan yang usianya di atas 30 tahun. Jika usia karyawan di atas 30, berikan peningkatan tambahan sebesar 2% dari gaji saat ini menggunakan fungsi lambda.

```python
data_tambah=[]
for u in zip(data_jumlah['Usia'],data_jumlah['Gajih_baru']): 
    if u[0]>30:
        tambahan = lambda u: u[1] + (0.02) * u[1]
        hasil_tambahan = tambahan(u)
        
    else:
        hasil_tambahan = "TETAP"
    
    data_tambah.append(hasil_tambahan)
    
data_jumlah_tambah= {'Usia': [25, 35, 30, 28],
            'Nama': ['John', 'Jane', 'Bob', 'Alice'],
            'Gajih_baru' :hasil_tamp,
            'Total_gaji_usia' :data_tambah}
            
data_jumlah_tambah= pd.DataFrame(data_jumlah_tambah)
data_jumlah_tambah
```
