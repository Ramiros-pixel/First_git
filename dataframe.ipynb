```python
import pandas as pd 

df = pd.read_csv('static\download\data.csv')
df
```

```python
tahun= 2023
total_sampah=0
for a,row in df.iterrows():
    if row['tahun'] == tahun:
        total_sampah += row['jumlah_sampah']
print(f'Total sampah pada tahun {tahun} adalah {total_sampah} ton')
```

```python
total_per_tahun = {}

for index, row in df.iterrows():
    tahun = row['tahun']
    jumlah = row['jumlah_sampah']

    if tahun not in total_per_tahun:
        total_per_tahun[tahun] = 0

    total_per_tahun[tahun] += jumlah

for tahun, total in total_per_tahun.items():
    print(f"Total sampah tahun {tahun}: {total}")
```

```python
kotanya={}

for index, row in df.iterrows():
    kota=row['nama_kabupaten_kota']


    if kota not in kotanya:
        kotanya[kota]=0 
    kotanya[kota] += row['jumlah_sampah']

for a,b in kotanya.items():
    print(f"kotanya adalah {a} jumlah sampah {b}")
# print(f"kotanya adalah {kota} jumlah sampah {kotanya}")
```
