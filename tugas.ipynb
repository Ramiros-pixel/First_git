17.	Buat kolom baru bernama kategori_dm dengan aturan berbasis jumlah_penderita_dm:
o	< 50.000 â†’ "Rendah"
o	50.000â€“99.999 â†’ "Sedang"
o	>= 100.000 â†’ "Tinggi"


```python
import numpy as np

kondisi = [
    df['jumlah_penderita_dm'] < 50000,
    (df['jumlah_penderita_dm'] >= 50000) & (df['jumlah_penderita_dm'] < 100000),
    df['jumlah_penderita_dm'] >= 100000
]

pilihan = ['Rendah', 'Sedang', 'Tinggi']

df['kategori_dm'] = np.select(kondisi, pilihan, default='Tinggi')
df[['jumlah_penderita_dm','kategori_dm']]
```

18.	Untuk setiap tahun, hitung persentase jumlah penderita DM per kabupaten/kota terhadap total jumlah penderita DM di tahun tersebut, lalu simpan dalam kolom baru bernama persentase_tahun.

20.	Buat grafik bar yang menampilkan jumlah_penderita_dm untuk setiap kabupaten/kota pada tahun 2019.
o	Sumbu-x: nama_kabupaten_kota
o	Sumbu-y: jumlah_penderita_dm
o	Beri judul dan label sumbu yang jelas.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Filter data tahun 2019

# Buat bar chart
plt.figure(figsize=(12, 6))
plt.bar(data_2019['nama_kabupaten_kota'], data_2019['jumlah_penderita_dm'])

# Label dan judul
plt.xlabel('Kabupaten/Kota')
plt.ylabel('Jumlah Penderita DM')
plt.title('Jumlah Penderita DM per Kabupaten/Kota Tahun 2019')

# Rotasi label x agar tidak overlap
plt.xticks(rotation=45, ha='right')



plt.show()
```

21.	Buat grafik garis (line chart) yang menampilkan total jumlah penderita DM Jawa Barat per tahun.
o	Sumbu-x: tahun
o	Sumbu-y: total jumlah_penderita_dm
o	Tambahkan titik data (marker) di setiap tahun.


```python

points= df.groupby('tahun')['jumlah_penderita_dm'].sum().reset_index()
plt.plot(points['tahun'], points['jumlah_penderita_dm'], marker='o')
plt.title("Jumlah penderita dm per Tahun di Jawa Barat")
plt.xlabel("Tahun")
plt.ylabel("Jumlah_penderita_dm")
plt.ticklabel_format(style='plain', axis='y')

plt.grid(alpha=0.3)

plt.show()
```

22.	Buat grafik bar horizontal untuk menampilkan 10 kabupaten/kota dengan jumlah_penderita_dm tertinggi pada tahun 2019.
o	Urutkan dari jumlah tertinggi ke terendah.


```python
tertinggi= df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum()
filt=tertinggi.sort_values(ascending=False).head(10)

plt.barh(filt.index, filt.values)
plt.xticks(rotation=45, ha='right')
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```

23.	Buat pie chart yang menunjukkan proporsi jumlah_penderita_dm berdasarkan kategori_dm ("Rendah", "Sedang", "Tinggi") pada tahun 2019.

```python

import numpy as np

# Hitung proporsi per kategori
proporsi = (data_2019.groupby('kategori_dm')['jumlah_penderita_dm'].sum())

plt.figure()
plt.pie(
    proporsi,
    labels=proporsi.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Proporsi Jumlah Penderita DM Berdasarkan Kategori Tahun 2019')
plt.show()
```
