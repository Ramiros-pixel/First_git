```python
import pandas as pd 

df = pd.read_csv('static/download/data.csv')
tahun2319= (df['tahun'] == 2023) | (df['tahun'] == 2018) 
jumlah_pertahun= df[tahun2319].groupby(df['nama_kabupaten_kota'])['jumlah_sampah'].sum()
median_pertahun=df[tahun2319].groupby(df['nama_kabupaten_kota'])['jumlah_sampah'].median()

tambah = pd.concat([jumlah_pertahun,median_pertahun],axis=1)
tambah.columns = ["jumlah tahunan","median tahunan"]

ini = tambah.reset_index()
ini
```
